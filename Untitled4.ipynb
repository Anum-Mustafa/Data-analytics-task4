```python
import pandas as pd 
import numpy as np

df = pd.read_csv("C:/Users/Rizwan Ahmed/Downloads/titanic/train.csv")

df

df.shape

# task statical summary of the data mean median and mode and std

df.describe()

df.isnull().sum()

# beacuse of numerical data in Age column we have to use mean or median for filling the missing values

avd = df['Age'].mean()

df['Age'] = df['Age'].fillna(avd)

df.isnull().sum()

df['Cabin']

df['Cabin'].unique()

df['Cabin'] = df['Cabin'].replace(np.nan, 0)

from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

# Fit the label encoder on the "Cabin" column and transform the values
df['Cabin'] = label_encoder.fit_transform(df['Cabin'].astype(str))

df['Cabin'].unique()

df

df.isnull().sum()

df['Embarked'] = df['Embarked'].dropna()

df

# sloved missing values task


# now outlier

import matplotlib.pyplot as plt 
import seaborn as sns

df['Age'].dtypes

df['Age'] = df['Age'].astype(int)

col2= df['Age']

sns.boxplot(col2)

import matplotlib.pyplot as plt

plt.title('Box Plot of Age')
plt.ylabel('Age Values')
plt.show()

col2.describe

q1=col2.quantile(0.25)
q3=col2.quantile(0.75)
print('percentile 25%', q1)
print('percentile 75%', q3)

iqr = q3 - q1
iqr

lower_limit= q1 - 1.5*iqr
upper_limit= q3 + 1.5*iqr
print('Lower Limit', lower_limit)
print('Upper Limit', upper_limit)

outliers_max= df[df['Age'] > upper_limit]
outliers_max

outliers_min= df[df['Age'] < lower_limit]
outliers_min

df[col2< lower_limit].count()

df[col2> upper_limit].count()

new_df= df[(col2> lower_limit)&(col2< upper_limit)]
new_df.shape

new_df

sns.boxplot(new_df['Age'])

col3 = new_df['Age']

sns.boxplot(col3)

```
